```python
#reading training data
#calling for generator  ###training in one set###
sys.path.append(os.path.abspath('/home/tc768/project/HKH/DL_seg_regression/code/dataPrepare'))
from dataGener1 import DataGenerator
#from dataGener_backup import DataGenerator
#changeable para
patch_shape = (32, 32, 7)
batch_size = 8
nband=7
#classNumber = 4 #change here
#patch folder
folderData = '/home/tc768/project/HKH/DL_seg_regression/patch/patch_32_b7_augalls2r3/' #not including snow

#number of training and validate samples
mat = sio.loadmat(folderData+'patchNum.mat')

patchNum=mat['patchNum']*1

traNum=np.sum(patchNum[0,:]) ;
print('train:', traNum)

#validata samples
fileVal = glob2.glob(folderData+'vali/' +'*.h5')
shuffle(fileVal)

#training samples
fileTra = glob2.glob(folderData+'trai/' +'*.h5')
shuffle(fileTra)
print('train files:', len(fileTra))

# Generators
params = {'dim_x': patch_shape[0],
		  'dim_y': patch_shape[1],
		  'dim_z': patch_shape[2],
		  'batch_size':batch_size,
		  'band':list(range(nband)), #change here
		  'flow': 0}


tra_MYgenerator = DataGenerator(**params).MYgenerate(fileTra)
val_MYgenerator = DataGenerator(**params).MYgenerate(fileVal)
print('training img',tra_MYgenerator[0].shape) #X
print('training label',tra_MYgenerator[1].shape) #y(label)
print('generator x dtype', tra_MYgenerator[0].dtype)
print('generator y dtype', tra_MYgenerator[1].dtype)

train_images = tra_MYgenerator[0]
train_masks = np.double(tra_MYgenerator[1])
train_masks = np.squeeze(train_masks) #reduce dim
val_images = val_MYgenerator[0]
val_masks = np.double(val_MYgenerator[1])
val_masks = np.squeeze(val_masks) #reduce dim
# train_masks[train_masks>classNumber] = 0 #why there are out of range value
# train_masks[train_masks<0] = 0
print('label dtype', train_masks.dtype)
print(np.max(train_masks))
print(np.min(train_masks))

frequencies=np.zeros(11, dtype= np.int64)
frequencies_val=np.zeros(11, dtype= np.int64)

for c in np.arange(0,11): 
	idx_ = np.where((train_masks > (c-1)/10)&(train_masks <= c/10))
	idx  = idx_[0]
	frequencies[c]=idx.shape[0]
	idx_ = np.where((val_masks > (c-1)/10)&(val_masks <= c/10))
	idx  = idx_[0]
	frequencies_val[c]=idx.shape[0]
    
    
    
print(frequencies)
print(frequencies_val)

```

```python
############transform dataset###############
#Encode labels... but multi dim array so need to flatten, encode and reshape
print(train_masks.shape)

#print(np.unique(train_masks_encoded_original_shape))
train_masks_input = np.expand_dims(train_masks, axis=3)
val_masks_input = np.expand_dims(val_masks, axis=3)
print(train_masks_input.shape)
print(train_masks_input.dtype)
#print(train_masks_input[1,:,:,:])
#print(np.max(train_masks_input))
X_train1=train_images
X_test1=val_images
```

```python
###############loss function and learning rate##################################

from tensorflow import keras
from tensorflow.keras.utils import normalize
from tensorflow.keras.metrics import MeanIoU

# Segmentation models losses can be combined together by '+' and scaled by integer or float factor
# set class weights for dice_loss (car: 1.; pedestrian: 2.; background: 0.5;)

import segmentation_models as sm
sm.set_framework('tf.keras')
sm.framework()

activation='sigmoid' #keep it within 0 and 1
batch_size = 8#8
optim = keras.optimizers.Adam(learning_rate=0.001)

def ssim_loss(y_true, y_pred):
  return 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1, filter_size=11))#half step

```

```python


epochs = 300
BACKBONE1 = 'resnet18'
file0='/home/tc768/project/HKH/DL_seg_regression/results/p32s16b7_augalls2r3/'#change here


timeCreated =time.strftime('%Y-%m-%d-%H', time.localtime(time.time()))
fileSave=  file0 + 'LR001'+ '_' +"ssim_2VTsplit" + '_' + str(batch_size)+'_' #smallerl kernel for ssim



preprocess_input1 = sm.get_preprocessing(BACKBONE1)

# define model
model1 = sm.Unet(BACKBONE1, input_shape=(None, None, 7), encoder_weights=None, activation=activation)
model1.compile(optimizer=optim, loss=ssim_loss, metrics=['mse','mae',ssim_loss]) #mae=mean absolute error


print(model1.summary())

print('X_train1',X_train1.shape)
print('y_train_cat',train_masks.shape)

##@markdown
checkpoint = ModelCheckpoint(fileSave+'solution1'+"_"+BACKBONE1+'_'+"weights.bestlr.hdf5", monitor='val_loss', 
                             verbose=1, save_best_only=True, save_weights_only=True, mode='auto', period=1)

tbCallBack = TensorBoard(log_dir=file0+'logs' + '_' + BACKBONE1 + '_' + str(batch_size) + '_' + "LRdecay" + '_' + timeCreated,  
                 histogram_freq=0, 
                 write_graph=True,
                 write_grads=True,
                 write_images=True,
                 embeddings_freq=0,
                 embeddings_layer_names=None,
                 embeddings_metadata=None)



history1=model1.fit(X_train1, 
          train_masks_input,
          batch_size=8, 
          epochs=epochs,
          verbose=1,
          validation_data=(X_test1, val_masks_input),
          #validation_split=0.25,
          callbacks = [checkpoint, tbCallBack])

plot_model(model1, to_file=file0+'model_plot_'+BACKBONE1+'.png', show_shapes=True, show_layer_names=True)
```

```python
#Reload model and continue training
BACKBONE1 = 'resnet18'
model1 = sm.Unet(BACKBONE1, input_shape=(None, None, 7), encoder_weights=None, activation=activation)
model1.load_weights('/home/tc768/project/HKH/DL_seg_regression/results/p32s16b7_augallsr/LR001_ssim_2VTsplit_8_solution1_resnet18_weights.bestlr.hdf5')
model1.compile(optimizer=optim, loss=ssim_loss, metrics=['mse','mae',ssim_loss]) #mae=mean absolute error

#Continue training
history2=model1.fit(X_train1, 
          train_masks_input,
          batch_size=8, 
          epochs=60,
          verbose=1,
          validation_data=(X_test1, val_masks_input),
          #validation_split=0.25,
          callbacks = [checkpoint, tbCallBack])


```
